## 3. Methodology

### Business Understanding

### Expolaroty Data Analysis

1-Demographics of Berlin: 
    -To get a list of demographics of Berlin, the data is extracted from Wikipedia page titled **Demographics of Berlin**     ("https://en.wikipedia.org/wiki/Demographics_of_Berlin").
    
   Berlin geopgraphical and demographics data,required data is scraped from the webpages by using BeautifulSoup and Requests packages of Python.
   
2-In order to use Foursquare API, we fetch the location data of all these neighborhoods from Python Geocoder package.

3-Afterthat, the Foursquare AOPI is used to get the top 50 venues in a raidus of 1500 meters of a particular borough.

4-To prepare the data for K-means clustering, we group the data frame by neighborhoods.

6-Finally, K-means clustering in performed on this data set to return 4 clusters,or categories of boroughs in terms of number of Italian Restaurants.
